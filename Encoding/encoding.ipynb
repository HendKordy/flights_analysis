```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('df_encoding.csv')
df.shape
```

```python
df.head()
```

```python
df["destination"].replace({"New Delhi": "Delhi"}, inplace=True)
df.head()
```

```python
df.embarked = df['embarked'].map('source_{}'.format)
df.head()
```

```python
df.destination = df['destination'].map('destination_{}'.format)
df.head()
```

```python
dummies1 = pd.get_dummies(df.airline)
dummies1
```

```python
dummies1.isnull().sum()
```

```python
dummies2 = pd.get_dummies(df.destination)
dummies2
```

```python
dummies2.isnull().sum()
```

```python
dummies3 = pd.get_dummies(df.embarked)
dummies3
```

```python
dummies3.isnull().sum()
```

```python
dummies4 = pd.get_dummies(df.stops)
dummies4
```

```python
dummies4.isnull().sum()
```

```python
dummies5 = pd.get_dummies(df.departure_time)
dummies5
```

```python
dummies5.isnull().sum()
```

```python
merged = pd.concat([df,dummies1,dummies2,dummies3,dummies4,dummies5],axis='columns')
merged
```

```python
df.head()
```

```python
merged.to_csv(r'df_encoding.csv', index = False)
```
