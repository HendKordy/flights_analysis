```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('df_cleaned.csv')
df.shape
```

```python
df.head()
```

```python
df["destination"].replace({"New Delhi": "Delhi"}, inplace=True)
df.head()
```

```python
df.embarked = df['embarked'].map('source_{}'.format)
df.head()
```

```python
df.destination = df['destination'].map('destination_{}'.format)
df.head()
```

```python
dummies1 = pd.get_dummies(df.airline)
dummies1
```

```python
dummies2 = pd.get_dummies(df.destination)
dummies2
```

```python
dummies3 = pd.get_dummies(df.embarked)
dummies3
```

```python
dummies4 = pd.get_dummies(df.stops)
dummies4
```

```python
df.departure_period = df['departure_period'].map('departure_{}'.format)
df.head()
```

```python
dummies5 = pd.get_dummies(df.departure_period)
dummies5
```

```python
df.arrival_period = df['arrival_period'].map('arrival_{}'.format)
df.head()
```

```python
dummies6 = pd.get_dummies(df.arrival_period)
dummies6
```

```python
merged = pd.concat([df,dummies1,dummies2,dummies3,dummies4,dummies5,dummies6],axis='columns')
merged
```
