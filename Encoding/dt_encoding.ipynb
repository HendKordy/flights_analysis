```python
import pandas as pd
```

```python
df = pd.read_csv('./df_encoding_test.csv')
```

```python
week_dummies = pd.get_dummies(df.weekday)
week_dummies.rename(columns={0: 'Monday', 1:'Tuesday', 2:'Wednesday', 3 : 'Thursday', 4 : 'Friday', 5: 'Saturday', 6:'Sunday'}, inplace=True)
```

```python
df = pd.concat([df, week_dummies], axis='columns')
```

```python
df.drop(['airline', 'embarked', 'destination', 'Route', 'stops', 'journey_year', 'weekday', 'arrival_period', 'departure_period', 'duration_hours', 'duration_mins', 'additional_info'], axis=1, inplace=True)
```

```python
df['Arrived_next_day'] = df['Arrived_next_day'].astype(int)
```

```python
df.sample(10)
```

```python
df.to_csv('../decision_tree/dt_encoded_test.csv', index=False)
```
